```python
#rename salary to monthly_salary
users_data.rename(
    columns={"Salary":"Monthly_Salary"},inplace=True)
```

```python
#show users data with  salary renamed as monthly_salary
users_data
```

```python
#rename City to Location using mapper
column_mapper={"City":"Location"}
users_data.rename(
    mapper=column_mapper,
    axis=1,
    inplace=True)
```

```python
#show users data with  City renamed as  Location 
users_data
```

```python
#add name_city column
users_data["Name_Location"]=users_data["Name"]+","+users_data["Location"]
#show results with name_city column added
users_data
```

```python
#drop name_location column
users_data.drop(columns=["Name_Location"],inplace=True)
#show users data with name_location deleted
users_data
```

```python
#fetch users from Addis Ababa
users_from_addis_ababa=users_data[users_data['Location']=='Addis Ababa']
#reset index
users_from_addis_ababa=users_from_addis.reset_index(drop=True)
#use 1 as label instead of 0 as default
users_from_addis_ababa.index=range(1,len(users_from_addis)+1)
#show results
users_from_addis_ababa
```

```python
#fetch users whose salary is greater than 250000
salary_more_than_250000=users_data[users_data['Monthly_Salary']>250000]
#reset index
salary_more_than_250000.reset_index(drop=True)
#use 1 as label
salary_more_than_250000.index=range(1,len(salary_more_than_250000)+1)
#show results
salary_more_than_250000
```

```python
#sort by age
#from youngest to eldest
users_data.sort_values(by=["Age"],inplace=True)
#reset index
users_data.index=range(1,len(users_data)+1)
#show results
users_data
```

```python
#select 1,5,10 row labels and columns labelled Location,Occupation,Monthly_Salary
#iloc selects all including the labels themselves
users_data.loc[[1,5,10,15],["Location","Occupation","Monthly_Salary",]]
```

```python
#selects items  using positions
# iloc exludes end position
users_data.iloc[[0,3,6],[0,6]]
```

```python
#sort data by salary
#sort from highest paid to least paid
users_data.sort_values(by=["Salary"],inplace=True,ascending=False)
#reset index
users_data.index=range(1,len(users_data)+1)
#show results
users_data
```

```python
#loc selects items using column names
#includes end position
#select first 5 rows from users_data
#select only name and salary columns
users_data.loc[0:5,["Name","Salary"]]
```

```python
#select first 5 rows
#select from name to salary columns range
users_data.loc[0:5,"Name":"Salary"]
```

```python
#use iloc
#exludes end position
#it uses column positions to select items
#select only first 5 rows
#and only first 5 columns
users_data.iloc[0:5,0:5]
```
